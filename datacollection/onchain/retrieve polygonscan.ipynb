```python
import requests
import pandas as pd
from pandas import json_normalize 
import numpy as np
import time
import random
from pathlib import Path
```

```python
def getTransactions(address, chain, authToken):
    while True:
        try:
            response = requests.get("{}/api?module=account&action=tokennfttx&address={}&startblock=0&endblock=99999999&page=1&offset=0&sort=asc&apikey={}".format(chain, address, authToken))
            response = response.json()
            if(response["message"] == 'OK'):
                return(json_normalize(response, "result"))
            elif(response["message"] == "No transactions found"):
                return None
            elif(response["message"] == "NOTOK"):
                print("NOTOK", address)
                time.sleep(random.randrange(5,15))
            else:
                print(response["message"], address)
        except Exception as err:
            print("Something went wrong:",str(err))
            time.sleep(random.randrange(5,15))

def getEtherscanTransactions(address):
    return(getTransactions(address, "https://api.etherscan.io", "U8PWV5QB1FSR5RVDFWZ5ABKWZ21WE1ME9S"))

def getPolygonTransactions(address):
    return(getTransactions(address, "https://api.polygonscan.com", "EYGTZ95VB8ND48X2J5H3A8DQSHCPZWF6EZ"))
```

```python
positions = pd.read_csv("/home/everything/Documents/Eigene Paper/work in progress/wip2022measuremeta/data/decentraland-positions-2022-10-03.csv")
```

```python
addresses = np.sort(positions.address.dropna().unique())
```

```python
address_chunks = np.array_split(addresses, 20)
```

```python
from pqdm.processes import pqdm

for idx, chunk in enumerate(address_chunks):
    filename = "ether_nft_part{}.csv".format(idx)
    path = Path(filename)
    if(path.is_file()):
        print("file", filename, "already exists, skipping")
        continue
    result_part = pqdm(chunk, getEtherscanTransactions, n_jobs=3)
    result_part = pd.concat(result_part)
    result_part.to_csv(filename, index=False)
```
